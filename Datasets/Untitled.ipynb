```python
import pandas as pd
import numpy as np
df2 = pd.DataFrame(np.random.randn(20, 6), columns=list('ABCDEF'))
df2
```

```python
from sklearn.preprocessing import StandardScaler

x = df2.values

x = StandardScaler().fit_transform(x)
pd.DataFrame(x)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2, svd_solver="arpack", random_state=5)
principalComponents = pca.fit_transform(x)
principalDf = pd.DataFrame(data = principalComponents)
principalDf
```
